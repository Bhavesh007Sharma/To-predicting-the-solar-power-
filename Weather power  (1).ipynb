```python
from sklearn import metrics
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import seaborn as sns
import catboost as ctb
model = ctb.CatBoostRegressor(iterations=10000,learning_rate=0.01)
model.fit(X_train, y_train)
print(); print(model)
    # make predictions
expected_y  = y_test
predicted_y = model.predict(X_test)
# summarize the fit of the model
print(); print(metrics.r2_score(expected_y, predicted_y))
print(); print(metrics.mean_squared_error(expected_y, predicted_y))
plt.figure(figsize=(10,10))
sns.regplot(expected_y, predicted_y, fit_reg=True, scatter_kws={"s": 100})
```

```python
from scipy.stats import pearsonr

# calculate Pearson's correlation
corr, _ = pearsonr(expected_y, predicted_y)
print('Pearsons correlation: %.3f' % corr)
```
